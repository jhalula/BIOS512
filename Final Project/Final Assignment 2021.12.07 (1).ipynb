```r
library('tidyverse')
```

```r
df = read_csv('https://data.cdc.gov/resource/8pt5-q6wp.csv?$limit=10000', col_types = cols(phase = col_character()))
```

```r
df 
```

```r
df_race_ethnicity = df %>%
    filter(group == 'By Race/Hispanic ethnicity',
          indicator == 'Symptoms of Depressive Disorder') 
```

```r
df_race_ethnicity
```

```r
options(repr.plot.width = 8)

p = ggplot(df_race_ethnicity, aes(x = time_period_start_date, y = value, color = subgroup)) +
    geom_line() +
    facet_wrap(~subgroup) +
    labs(x = "Start Time for Collection Period", y = "Symptoms of Depressive Disorder", color = "Race/Etnicity") +
    scale_color_brewer(palette = "Dark2") +
    theme_minimal() +
    theme(strip.text = element_text(size = 5))

p
```

Figure 1 shows symptoms of depressive disorder across time of COVID 19, facetted by race/ethnicity. At the beginning of the data collection period, participants were reporting an increase in symptoms of depressive disorder as time during the COVID 19 pandemic progressed. Following the initial increase in symptoms of depressive disorder with the start of the Covid 19 pandemic, symptom values reduced as participants became more acclimated to COVID 19. 

```r
df_state = df %>%
    filter(indicator == 'Symptoms of Depressive Disorder',
          group == 'By State',
          time_period_start_date == '2020-04-23') %>%
    group_by(state) %>%
    summarize(value = sum(value)) %>%
    arrange(value) %>%
    mutate(state = fct_inorder(state))
```

```r
options(repr.plot.width = 10, repr.plot.height = 15, repr.plot.res = 250)

p = ggplot(df_state, aes(x = state, y = value)) +
    geom_col() +
    theme_minimal() +
    coord_flip() +
    labs(x = "State", y = "Symptoms of Depressive Disorder")
p
```

Figure 2 depicts United States' states and their associated symptoms of depressive disorder values for the time period of April 23 to May 5, 2020. The states are organized in descending order by value of symptoms of depressive disorder. It is interesting to note that there are not any obvious trends to the ranking, such as political leaning of the state, geographic location, or population density. 

```r
df_age = df %>%
    filter(group == 'By Age',
          indicator == 'Symptoms of Anxiety Disorder',
          time_period_end_date == '2020-05-05')
```

```r
p = ggplot(df_age, aes(x = subgroup, y = value, fill = subgroup)) +
    geom_col() +
    theme_minimal() +
    scale_color_brewer(palette = "Set2") +
    labs(x = "Age Range", y = "Symptoms of Anxiety Disorder", fill = "")

p    
```

Figure 3 depicts reported symptoms of anxiety disorder across different age ranges for United States residents for the time period of April 23 to May 5, 2020. Symptoms of anxiety were inversely related to age during this time period.
