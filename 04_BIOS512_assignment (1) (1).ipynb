```r
library('tidyverse')
```

```r
wind_turbine_df = read_csv(
    'https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-10-27/wind-turbine.csv',
    col_types = cols()
    )
```

```r
wind_turbine_df %>% head
```

Each row in the table represents a single turbine.

Make a bar chart üìä **sorted by frequency** of turbine count per province territory. (Hint: `fct_infreq` will sort your factor levels by frequency for subsequent plotting)

```r
df_sorted = wind_turbine_df %>%
    mutate(province_territory_sorted = province_territory %>% fct_infreq() %>% fct_relevel('Other', after = Inf) %>% fct_rev())

p = ggplot(df_sorted, aes(y = province_territory_sorted)) +
    geom_bar() +
    theme_minimal() +
    labs(x = "Count", y = "Province/Territory") +
    geom_text(stat='count', aes(label=..count..))

p
```

Make a bar chart **sorted by frequency** of turbine count per manufacturer. **Group the low frequency manufacturers into an `Other` category**.

**Bonus:** Can you force the `Other` category to be the last bar? 

```r
df_sorted_2 = wind_turbine_df %>%
    mutate(manufacturer_sorted = manufacturer %>% fct_lump(n = 6) %>% fct_infreq() %>% fct_relevel('Other', after = Inf) %>% fct_rev())
    

p = ggplot(df_sorted_2, aes(y = manufacturer_sorted)) +
    geom_bar() +
    theme_minimal() +
    labs(x = "Count", y = "Manufacturer") +
    geom_text(stat='count', aes(label=..count..))

p
```
